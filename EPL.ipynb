```python
#CHANGE FROM CATEGORICAL INTO NUMIRIC FOR SCORE FEATURE
result = []
Score_list=df.Score.str.split("-", expand = True)
Score_list=Score_list.astype(int)
Score_list.columns = ['home', 'away']
print(Score_list)
print(type(Score_list))

for index, row in Score_list.iterrows():
    #print(row['home'], row['away'])
    if row['home'] > row['away']:
        result.append(1) 
    elif row['home'] < row['away']:
       result.append(0)
    else:
        result.append(2)
df['Result'] = result
```

```python
clubs = {'MAN UTD':1, 'WEST BROM':2, 'LEICESTER CITY':3, 'WEST HAM':4, 'QPR':5,
       'STOKE':6, 'ARSENAL':7, 'LIVERPOOL':8, 'NEWCASTLE':9, 'BURNLEY':10,
       'ASTON VILLA':11, 'SWANSEA':12, 'CHELSEA':13, 'SOUTHAMPTON':14,
       'CRYSTAL PALACE':15, 'EVERTON':16, 'TOTTENHAM':17, 'HULL CITY':18,
       'SUNDERLAND':19, 'MAN CITY':20, 'NORWICH':21, 'BOURNEMOUTH':22, 'WATFORD':23,
       'MIDDLESBROUGH':24, 'BRIGHTON':25, 'HUDDERSFIELD':26, 'FULHAM':27, 'WOLVES':28,
       'CARDIFF':29, 'SHEFFIELD UTD':30, 'LEEDS UTD':31}
df['Home Team'] = df['Home Team'].map(clubs)
```
